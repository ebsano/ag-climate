```python
import requests

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Temperature in celcius
t_historical = pd.read_csv('historical_yearly_temperature.csv')
t_historical
```

```python
# rename the year column
t_historical = t_historical.rename(columns={'Unnamed: 0': 'year'}) 
t_historical
```

```python
# make the year column as index
t = t_historical.set_index('year')
t
```

```python
# make sure there is not much missing data
for col in t.columns:
    if t[col].isna().sum() > 5:
        print(col)
        print(f' {t[col].isna().sum()/t.shape[0]*100}% of data is missed.')
```

```python
# delete the countries/columns with high percentage of missing data
t = t.drop(columns=['Wake Island (U.S.)', 'serbia and montenegro'])
t
```

```python
# add yearly average (mean) temperature as new columns
t = t.assign(average= t.mean(axis=1))
t
```

```python
plt.figure()
plt.plot(t['average'])
plt.ylabel('global average temperature')
```

```python
# There might be many missing data for the last couple of years
t_test = t.assign(missed = t.isna().sum(axis=1))
t_test
# The guess was correct! For 39 out of 112 countries, the precipitation is not reported for years 2010-2012
```

```python
# Therefore, it is better to exclude those years
t = t.drop([2010,2011,2012])
t
```

```python
plt.figure()
plt.plot(t['average'])
plt.ylabel('global average temperature')
```
