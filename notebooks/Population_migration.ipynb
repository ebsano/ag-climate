```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import folium
from google.colab import drive
drive.mount('/content/drive/')
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/DS4A-W_Team25/Colab_Files/data/Total_Population.csv')
df1.info()
```

```python
df_pop = df1[df1['Year(s)']== 2019]
df_pop.info()
```

```python
m = folium.Map()

#Setting up the world countries data URL
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
country_shapes = f'{url}/world-countries.json'

folium.Choropleth(
    #The GeoJSON data to represent the world country
    geo_data=country_shapes,
    name='Population',
    data=df_pop,
    #The column accepting list with 2 value; The country name and  the numerical value
    columns=['Country or Area', 'Value'],
    key_on='feature.properties.name',
    fill_color='PuRd',
    nan_fill_color=''
).add_to(m)
m
```

```python
df3 = pd.read_csv('/content/drive/MyDrive/DS4A-W_Team25/Colab_Files/data/Net_Migration.csv')
df3.dropna(how='all', axis=1, inplace=True)
```

```python
df3.info()
```

```python
df3[df3['Country Name']=='Venezuela']
```

```python
df3.replace("United States", "United States of America", inplace = True)
df3.replace("Russian Federation", "Russia", inplace = True)
df3.replace("Venezuela, RB", "Venezuela", inplace = True)
df3.replace('Congo, Dem. Rep.', "Democratic Republic of the Congo", inplace = True)
df3.replace("Cote d'Ivoire", "Ivory Coast", inplace = True)
df3.replace('Congo, Rep.', "Republic of the Congo", inplace = True)
df3.replace('Egypt, Arab Rep.', "Egypt", inplace = True)
df3.replace('Iran, Islamic Rep.', "Iran", inplace = True)
df3.replace('Tanzania',"United Republic of Tanzania", inplace = True)
df3.replace('Lao PDR', "Laos", inplace = True)
df3.replace('Syrian Arab Republic', "Syria", inplace = True)
df3.replace('Serbia', "Republic of Serbia", inplace = True)
df3.replace('Yemen, Rep.', "Yemen", inplace = True)
df3.replace("Korea, Dem. People's Rep.", "South Korea", inplace = True)
df3.replace("Kyrgyz Republic", "Kyrgyzstan", inplace = True)
```

```python
df3['Country Name'].unique();
```

```python
m1 = folium.Map()

#Setting up the world countries data URL
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
country_shapes = f'{url}/world-countries.json'

folium.Choropleth(
    #The GeoJSON data to represent the world country
    geo_data=country_shapes,
    name='Net Migration',
    data=df3,
    #The column aceppting list with 2 value; The country name and  the numerical value
    columns=['Country Name', '2017'],
    key_on='feature.properties.name',
    fill_color='PuRd',
    nan_fill_color='white'
).add_to(m1)
m1
```
