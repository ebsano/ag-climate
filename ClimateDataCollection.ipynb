```python
url = 'http://climatedataapi.worldbank.org/climateweb/rest/'
```

```python
params = {
    'v1': 'v1',
    'country':'country',
    'type': 'annualavg', # annual average
    'var': 'pr', #precipitation
    'start': '1920',
    'end': '1939',
    'ISO3': 'afg'
}
```

```python
response = requests.get(url+'v1/country')
```

```python
country_codes = {}
for entry in response.json():
  country_codes[entry['name']] = entry['iso3']
```
